# Exercise 2 The law of large numbers

The law of large numbers states that a relative frequency will stabilize in the limit of large
numbers (respectively innitey). Write a short numpy code that simulates N coin tosses and
computes the cumulative relative frequency at each subsequent coin toss. Visualize this cumulative relative frequency. Does your simulation stably reach 0.5 in the limit of large N? How
large does N approximately need to be to reach at least three zeros behind the 0.5? (or three
nines behind the 0.4)


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
intN = 10_000
listRes = np.random.randint(0, 2, size=intN) # just like with the range function, the end isn't part of the possiblities
listRes[:20]
```

```python
listN = np.arange( 1, intN+1, 1) # A list from 1 until intN
listProbs = np.cumsum( listRes ) / listN # np.cumsum returns a list of len(N) where each element is added on top of the last one
listProbs[:20] # listProbs consinsts of cumulative sum of numbers divided by total amount of tosses N
```

```python
print(listN)
print(np.cumsum(listRes))
```

```python
plt.figure(figsize=(15, 5))
plt.plot(listN, listProbs, lw=2)
plt.xlabel("Tosses", fontsize=20)
plt.ylabel("Probability", fontsize=20)
plt.axhline(y=0.5, c="black", lw=1, ls="--")
```

```python
listProbs[-1]
```

With just 10'000 tosses we're way off from 0.500 or anything similar.
